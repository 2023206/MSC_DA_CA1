```python
male_total = dist_population.loc[dist_population['Sex'] == 'Male', 'VALUE'].sum()
female_total = dist_population.loc[dist_population['Sex'] == 'Female', 'VALUE'].sum()

#Z-test: in this test I can compare the total of the population of men and women to determinate
#if the difference is significant and it is worthwhile treating the groups separately
z_stat, p_value = stats.ttest_ind(dist_population.loc[dist_population['Sex'] == 'Male', 'VALUE'], dist_population.loc[dist_population['Sex'] == 'Female', 'VALUE'])

alpha = 0.05  # tipycal value for this kind of test
print("Total for Male:", male_total)
print("Total for Female:", female_total)
print("z-statistic:", z_stat)
print("p-value:", p_value)

if p_value < alpha:
    print("There is a significant difference in the number of males and females.")
else:
    print("No significant difference in number of men and women.")

#In my case I consider that the variables in sex are related, since the increase of one represents the decrease of the 
#other in the general population. In the research I conducted, the "P-value" measures the probability of obtaining a value
#that the one obtained in the database study and "ALPHA" is a standard value in the equation that measures the significance
#level, it is the probability that I make the mistake of rejecting the null hypothesis when in fact it is true.
#Since my null hypothesis is that the difference between the totals is significant and knowing that if the p-value is 
#greater than alpha, the null hypothesis should be rejected. So I can conclude that 
#the difference is not large enough to be a factor in the investigation. 
```

```python
age_groups = ['0 - 4 years', '5 - 9 years', '10 - 14 years', '15 - 19 years',
 '20 - 24 years', '25 - 29 years', '30 - 34 years', '35 - 39 years',
 '40 - 44 years', '45 - 49 years', '50 - 54 years', '55 - 59 years',
 '60 - 64 years', '65 - 69 years', '70 - 74 years', '75 - 79 years',
 '80 - 84 years', '85 years and over']

age_dict = {age_groups[i]: i for i in range(len(age_groups))}

for group, group_id in age_dict.items():
    print(f"{group}: {group_id}")
#in this step I created a dictionary to assign a numerical value to each age group, 
#as this will allow me in later steps to relate this variable with other numerical variables
df_filtered['Encoded Age Group'] = df_filtered['Age Group'].map(age_dict)
```

```python
group_ranges = {
    "0-19": [0, 1, 2, 3],
    "20-59": [4, 5, 6, 7, 8, 9, 10, 11],
    "60 and over": [12, 13, 14, 15, 16, 17, 18]
}

grouped_age_dict = {k: [] for k in group_ranges}

for group, group_id in age_dict.items():
    for key, value in group_ranges.items():
        if group_id in value:
            grouped_age_dict[key].append(group)

print(grouped_age_dict)
#I group the age ranges to reduce capillarity, this will help me in two ways, 
#the first will be to be able to treat the largest amount of the population 
#as a single group and the second will be to group the majority of high-risk people in another group
```

```python
grouped_age_dict
```

```python
grouped_value_totals = {k: 0 for k in grouped_age_dict}

for group, age_group in grouped_age_dict.items():
    for age in age_group:
        grouped_value_totals[group] += df_filtered[df_filtered['Age Group'] == age]['VALUE'].sum()

print(grouped_value_totals)
#determined the population in each new age group
```

```python
total = sum(grouped_value_totals.values())
percentages = {k: v / total * 100 for k, v in grouped_value_totals.items()}

fig, ax = plt.subplots()
ax.pie(percentages.values(), labels=percentages.keys(), autopct='%1.1f%%', startangle=90)
ax.axis('equal')

plt.show()

#chart for new age groups
```

```python
senior_group = grouped_age_dict["60 and over"]
df_seniors = df_filtered[df_filtered['Age Group'].isin(senior_group)]

seniors_by_county = df_seniors.groupby('County')['VALUE'].sum()

seniors_by_county = seniors_by_county.sort_values(ascending=False)

print(seniors_by_county.head(42))

#In this step I use my new age groups to count how many people over 60 years of age there are in each county 
#and order it in descending order, since one of my objectives in this study is to verify the distribution of people 
#who require greater care and attention, and I consider this group to be the most at risk, based on general health charts
```

```python
#To support my theory about risk, I define a function that associates the age range and general health; 
#the older and worse the health, the risk of death is considered greater

def assign_risk(row):
    age_group = row['Age Group']
    if any(age_group in group for group in grouped_age_dict.values()):
        if row['General Health'] in ['General health - Very good', 'General health - Good']:
            return 'Low Risk'
        elif row['General Health'] in ['General health - Fair', 'General health - Bad']:
            return 'Medium Risk'
        elif row['General Health'] == 'General health - Very Bad':
            return 'High Risk'
    else:
        return 'Not Applicable'

df_filtered['Risk'] = df_filtered.apply(assign_risk, axis=1)
print(df_filtered[['Age Group', 'General Health', 'Risk']])

#This function will be applied to each record in order to determine the number of people associated with each risk level
```

```python
#Function to add the column of the new age group to the df and complete it with the information 
#previously obtained when creating the groups

def assign_age_group(row):
    for age_group, age_group_values in grouped_age_dict.items():
        if row['Age Group'] in age_group_values:
            return age_group

df_filtered['grouped_age'] = df_filtered.apply(assign_age_group, axis=1)

print(df_filtered[['Age Group', 'grouped_age', 'General Health', 'Risk']])
```

```python
columns_to_keep = ['County','Encoded Sex', 'Encoded Age Group', 'Encoded General Health', 'Risk_Encoded', 'VALUE']

# Crear un nuevo DataFrame con las columnas seleccionadas
df_encoded = df_filtered[columns_to_keep].copy()
```

```python
features = ['Encoded Age Group', 'Encoded General Health']
X = df_encoded[features]
y = df_encoded['Risk_Encoded']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=3)

#clasificador
clf = DecisionTreeClassifier()

# Entrenar el modelo
clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

clf = clf.fit(X_train, y_train)

print("Accuracy of the tree:", metrics.accuracy_score(y_test, y_pred))

```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plot_tree(clf, feature_names=features, class_names=[str(x) for x in clf.classes_], filled=True, rounded=True)
plt.title("Decision Tree")
plt.show()
```
